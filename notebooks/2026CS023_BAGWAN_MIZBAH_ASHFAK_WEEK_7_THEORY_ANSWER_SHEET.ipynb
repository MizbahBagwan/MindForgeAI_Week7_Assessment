## Q1 · Data splitting and leakage · 5 marks

Distinguish training, validation and test data. Explain two forms of data leakage and show how a pipeline prevents one of them.

Training Data: Used to train the machine learning model.

Validation Data: Used to tune the model and select the best one.

Test Data: Used at the end to check the final performance on new data.

Example:

Training = 70%
Validation = 15%
Test = 15%

Two types of data leakage:

Target Leakage: Using future information.
Example: Using the final exam marks (G3) to predict student risk before the exam.
Preprocessing Leakage: Applying scaling or encoding to the whole dataset before splitting.
Example: Standardizing all student records before creating train and test sets.

How Pipeline helps:

A pipeline applies preprocessing only on the training data and then applies the same transformation to validation and test data, preventing data leakage.

## Q3 · Cross-validation strategy · 5 marks

Explain K-fold cross-validation. State when ordinary K-fold is invalid or weak, and select suitable alternatives for (a) imbalanced classes, (b) repeated customers/devices and (c) time-ordered prediction.

Your Answer

K-Fold Cross Validation divides the dataset into K equal parts.

The model is trained on K−1 folds and tested on the remaining fold.

This process repeats until every fold has been used once.

Example:

If K = 5, the data is divided into 5 parts.

Round 1 → Train on 4 folds, Test on 1 fold
Round 2 → Train on another 4 folds, Test on the next fold
Continue until all 5 folds are tested.

Different methods:

Imbalanced data: Stratified K-Fold
Repeated customers/devices: Group K-Fold
Time-based data: Time Series Split

Overfitting: The model memorizes the training data and performs poorly on new data.

Example: A model gets 100% training accuracy but only 70% test accuracy.

Ways to improve:
Add useful features.
Remove unnecessary features.
Use cross-validation.
Use regularization or early stopping.

Epoch:
An epoch means the model has gone through the entire training dataset once.

Example: If there are 1000 student records, then after reading all 1000 records one time, it is called 1 epoch.

Iteration:
An iteration is one step of training. There can be many iterations in one epoch.

One-call model fitting:
Some machine learning algorithms, like Decision Tree and Random Forest, are trained using only one fit() function. They do not use epochs.

Early Stopping:
It stops the training when the model is no longer improving.

Patience:
Patience means how many epochs or iterations the model waits before stopping.

Convergence:
Convergence means the model has learned enough, and further training does not improve the results.
