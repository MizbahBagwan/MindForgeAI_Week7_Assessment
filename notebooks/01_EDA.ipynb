```python
import pandas as pd

# Raw data load
df = pd.read_csv("../data/raw/student-mat.csv")

# Cleaning
df = df.drop_duplicates()
df = df.dropna()

# Save cleaned data
df.to_csv(
    "../data/interim/cleaned_student_data.csv",
    index=False
)

print("Cleaned data saved ✅")
```

```python
import pandas as pd
import os


# Load cleaned interim data
df = pd.read_csv(
    "../data/interim/cleaned_student_data.csv"
)


# Create target variable
df["risk"] = (df["G3"] < 10).astype(int)


# Select important features
features = [
    "age",
    "studytime",
    "failures",
    "absences"
]


# Create model-ready dataset
processed_df = df[features + ["risk"]]


# Create processed folder
os.makedirs(
    "../data/processed",
    exist_ok=True
)


# Save processed data
processed_df.to_csv(
    "../data/processed/model_ready_data.csv",
    index=False
)


print("Processed data saved ✅")
print(processed_df.head())
print(processed_df.shape)
```
